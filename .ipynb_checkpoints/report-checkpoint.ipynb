```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get(path_or_url):
    return pd.read_csv(path_or_url)
```

```python
def plot_profiling(dataframe_lhs, dataframe_rhs):
    # plot
    xsize = 15
    ysize = 5    
    max_xlabels = 50 
    
    fig = plt.figure(figsize=(xsize, ysize))
    ax  = plt.axes()
    ax.xaxis.set_major_locator(plt.MaxNLocator(max_xlabels))
    
    plt.xticks(rotation=90)
    plt.xlabel("size")
    plt.ylabel("time")
    plt.grid()
    
    ax.plot(dataframe_lhs.tree_size, dataframe_lhs.insert_time, label="insert avl")
    ax.plot(dataframe_lhs.tree_size, dataframe_lhs.find_time, label="find avl")
    ax.plot(dataframe_lhs.tree_size, dataframe_lhs.erase_time, label="erase avl")
    
    ax.plot(dataframe_rhs.tree_size, dataframe_rhs.insert_time, label="insert rb")
    ax.plot(dataframe_rhs.tree_size, dataframe_rhs.find_time, label="find rb")
    ax.plot(dataframe_rhs.tree_size, dataframe_rhs.erase_time, label="erase rb")
    
    plt.legend(loc="upper left")
    plt.show()    
```

```python
avl = get("profiler/results/avl.csv")
rb = get("profiler/results/set.csv")
```
