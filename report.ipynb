```python
def get(path_or_url):
    return pd.read_csv(path_or_url)
```

```python
def plot_profiling_insert(dataframe1, dataframe2, dataframe3, dataframe4):
    # plot
    xsize = 15
    ysize = 5    
    max_xlabels = 50 
    
    fig = plt.figure(figsize=(xsize, ysize))
    ax  = plt.axes()
    ax.xaxis.set_major_locator(plt.MaxNLocator(max_xlabels))
    
    plt.xticks(rotation=90)
    plt.xlabel("size")
    plt.ylabel("time")
    plt.grid()
    
    ax.plot(dataframe1.tree_size, dataframe1.insert_time, label="insert avl")
    ax.plot(dataframe2.tree_size, dataframe2.insert_time, label="insert splay")
    ax.plot(dataframe3.tree_size, dataframe3.insert_time, label="insert rb")
    ax.plot(dataframe4.tree_size, dataframe4.insert_time, label="insert cartesian")
    
    plt.legend(loc="upper left")
    plt.show()    
    
def plot_profiling_find(dataframe1, dataframe2, dataframe3, dataframe4):
    # plot
    xsize = 15
    ysize = 5    
    max_xlabels = 50 
    
    fig = plt.figure(figsize=(xsize, ysize))
    ax  = plt.axes()
    ax.xaxis.set_major_locator(plt.MaxNLocator(max_xlabels))
    
    plt.xticks(rotation=90)
    plt.xlabel("size")
    plt.ylabel("time")
    plt.grid()
    
    ax.plot(dataframe1.tree_size, dataframe1.find_time, label="find avl")
    ax.plot(dataframe2.tree_size, dataframe2.find_time, label="find splay")
    ax.plot(dataframe3.tree_size, dataframe3.find_time, label="find rb")
    ax.plot(dataframe4.tree_size, dataframe4.find_time, label="find cartesian")
    
    plt.legend(loc="upper left")
    plt.show()  
    
def plot_profiling_erase(dataframe1, dataframe2, dataframe3, dataframe4):
    # plot
    xsize = 15
    ysize = 5    
    max_xlabels = 50 
    
    fig = plt.figure(figsize=(xsize, ysize))
    ax  = plt.axes()
    ax.xaxis.set_major_locator(plt.MaxNLocator(max_xlabels))
    
    plt.xticks(rotation=90)
    plt.xlabel("size")
    plt.ylabel("time")
    plt.grid()
    
    ax.plot(dataframe1.tree_size, dataframe1.erase_time, label="erase avl")
    ax.plot(dataframe2.tree_size, dataframe2.erase_time, label="erase splay")
    ax.plot(dataframe3.tree_size, dataframe3.erase_time, label="erase rb")
    ax.plot(dataframe4.tree_size, dataframe4.erase_time, label="erase cartesian")    
    
    plt.legend(loc="upper left")
    plt.show() 
    
    
```

```python
avl = get("profiler/results/avl.csv")
rb = get("profiler/results/set.csv")
splay = get("profiler/results/splay.csv")
cartesian = get("profiler/results/cartesian.csv")
```

```python
plot_profiling_insert(avl, splay, rb, cartesian)
```

```python
plot_profiling_find(avl, splay, rb, cartesian)
```

```python
plot_profiling_erase(avl, splay, rb, cartesian)
```
